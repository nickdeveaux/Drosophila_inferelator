```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import os
%matplotlib inline
import seaborn as sns
from scipy.stats import hypergeom
import scipy.stats as st
import numpy as np
import copy
import collections
```

```python
network_path = '/Users/ndeveaux/Dropbox (Simons Foundation)/HIV_Collaboration_2016/2017_activities/outlier_kept_in_full_model_100_subsamples/pval1_quartile_priors/network_1_to_399_by_betas_thresholded_at_75000_edges.tsv'

human_network= pd.read_csv(network_path, sep="\t")
```

```python
def get_dict_of_occurences(net):
    counts = {}
    for idx, row in net.iterrows():
        tf = row['regulator']
        if tf in counts:
            counts[tf] = counts[tf] + 1
        else:
            counts[tf] = 1
    return counts

def plot_dict(D):
    plt.bar(range(len(D)), list(sorted(D.values())), align='center')
    plt.show()
```

```python
network_path = '/Users/ndeveaux/Dev/inferelator_ng/data/yeast/2017-12-04_14-12-50/network.tsv'

yeast_network= pd.read_csv(network_path, sep="\t")
```

```python
dmel_folder = '/Users/ndeveaux/Dropbox (Simons Foundation)/Drosophila_Inferelator_Project/network_inputs_and_outputs/outputs'
dmel_net_shortlist= pd.read_csv(os.path.join(dmel_folder, '2018-06-13_13-41-39/network.tsv'), sep="\t")
dmel_net_longlist= pd.read_csv(os.path.join(dmel_folder, '2018-06-15_05-25-53/network.tsv'), sep="\t")
```

```python
human_occurences = get_dict_of_occurences(human_network)
yeast_occurences = get_dict_of_occurences(yeast_network)
```

```python
dmel_net_shortlist_occurences = get_dict_of_occurences(dmel_net_shortlist)
dmel_net_longlist_occurences = get_dict_of_occurences(dmel_net_longlist)
```

```python
plt.title('number of targets per TF, human dendritic cells')
plot_dict(human_occurences)
```

```python
plt.title('number of targets per TF, yeast network')
plot_dict(yeast_occurences)
```

```python
plt.title('number of targets per TF, dmel_shortlist network')
plot_dict(dmel_net_shortlist_occurences)
```

```python
plt.title('number of targets per TF, dmel_longlist network')
plot_dict(dmel_net_longlist_occurences)
```
